**Aim :** The aim of this notebook is to perform **Exploratory Data Analysis (EDA)** on the preprocessed dataset obtained from Notebook 1. This includes both graphical and non-graphical analysis to understand the structure, patterns, and relationships present in the data.

**Notebook-2**

**Step 1:-** Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

**Step 2:-** Load Dataset

```python
# Create a dummy train.json file for demonstration if it doesn't exist
import os

file_name = 'train.json'
if not os.path.exists(file_name):
    dummy_data = [
        {"sentence": "This is a test sentence.", "relation": "NoRelation"},
        {"sentence": "Another sample sentence for EDA.", "relation": "HasRelation"}
    ]
    with open(file_name, 'w') as f:
        json.dump(dummy_data, f, indent=4)
    print(f"'{file_name}' created with dummy data.")
else:
    print(f"'{file_name}' already exists. Skipping dummy data creation.")
```

```python
# Load JSON file
with open('train.json', 'r') as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.json_normalize(data)

# Select important columns
df = df[['sentence', 'relation']]

df.head()
```

**Explanation:**
* Dataset loaded from JSON
* Converted into DataFrame
* Selected relevant columns

# **NON-GRAPHICAL ANALYSIS**

**1. Univariate Analysis (Sentence Length)**

```python
df['sentence_length'] = df['sentence'].apply(len)
df['sentence_length'].describe()
```

**Objective:**

To understand the distribution of sentence lengths.

**Observation:**
* Mean sentence length = (value)
* Minimum and maximum lengths observed
* Data spread (std deviation)

**2. Multivariate Analysis (Relation Count)**

```python
df['relation'].value_counts()
```

**Objective:**

To analyze the frequency of different relations.

**Observation:**
* Some relations appear more frequently
* Dataset may be imbalanced

# **GRAPHICAL ANALYSIS**

**1. Histogram (Sentence Length)**

```python
plt.hist(df['sentence_length'])
plt.title("Sentence Length Distribution")
plt.xlabel("Sentence Length")
plt.ylabel("Frequency")
plt.show()
```

**Objective:**

To visualize distribution of sentence lengths.

**Observation:**
* Most sentences fall within a specific range
* Data may be slightly skewed

**2. Bar Chart (Relation Distribution)**

```python
df['relation'].value_counts().head(10).plot(kind='bar')
plt.title("Top Relations")
plt.xlabel("Relation")
plt.ylabel("Count")
plt.show()
```

Objective:

To visualize most frequent relationships.

**Observation:**
* Few relations dominate
* Some relations are rare

**3. Scatter Plot (Sentence Length vs Relation Count)**

```python
relation_counts = df['relation'].value_counts()
plt.scatter(range(len(relation_counts)), relation_counts.values)
plt.xlabel("Relation Index")
plt.ylabel("Frequency")
plt.show()
```

**Objective:**

To analyze variation in relation frequencies.

**Observation:**
* Large variation in counts
* Some relations have very high frequency

**Final Conclusion**

**In this notebook:**

* Analyzed sentence length and relation distribution
* Identified imbalance in dataset
* Visualized patterns using graphs
