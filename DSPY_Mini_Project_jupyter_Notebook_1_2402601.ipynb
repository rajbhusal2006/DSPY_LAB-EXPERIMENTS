**Step 2:-** Load Dataset

```python
import json

# Create dummy MultiTACRED-like data
dummy_data = [
    {
        "sentence": "Barack Obama was born in Honolulu, Hawaii.",
        "subj_start": 0,
        "subj_end": 11,
        "obj_start": 21,
        "obj_end": 28,
        "relation": "per:place_of_birth"
    },
    {
        "sentence": "Apple Inc. was founded by Steve Jobs.",
        "subj_start": 0,
        "subj_end": 9,
        "obj_start": 24,
        "obj_end": 33,
        "relation": "org:founded_by"
    },
    {
        "sentence": "The Eiffel Tower is located in Paris, France.",
        "subj_start": 4,
        "subj_end": 16,
        "obj_start": 30,
        "obj_end": 35,
        "relation": "loc:located_in"
    }
]

# Save the dummy data to 'train.json'
with open('train.json', 'w') as f:
    json.dump(dummy_data, f, indent=4)

print("A dummy 'train.json' file has been created.")
print("Now, please re-run the data loading cell (VK7KxgSDRn6_) above this cell.")
```

**Explanation:**

* Dataset is stored in **JSON format**.

* Converted into structured table using pandas.

**Step 3:-** Data Selection

The dataset consists of sentences with relationships between two entities.

**Selected Features:**
* Sentence (text)
* Subject (entity 1)
* Object (entity 2)
* Relation (target variable)

**Reason:**
* These attributes are important for relation extraction task.
* Irrelevant metadata is ignored.

**Step 4:-** Understand Dataset

```python
import pandas as pd
import json

# Load the dummy data from 'train.json' into a DataFrame
with open('train.json', 'r') as f:
    data = json.load(f)
df = pd.DataFrame(data)

df.info()
df.describe(include='all')
```

**Explanation:**
* **info()** → shows number of rows, columns, data types
* **describe()**→ statistical summary

**Step 5:-** Attribute Description

| Attribute  | Description                   | Data Type |
| ---------- | ----------------------------- | --------- |
| sentence   | Input text sentence           | object    |
| subj_start | Start position of subject     | int       |
| subj_end   | End position of subject       | int       |
| obj_start  | Start position of object      | int       |
| obj_end    | End position of object        | int       |
| relation   | Relationship between entities | object    |

**Step 6:-** Check Missing Values

```python
df.isnull().sum()
```

**Explanation:**
* Checks for missing or null values
* Ensures dataset completeness

**Step 7:-** Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

**Explanation:**
* Removes repeated records
* Improves model performance

**Step 8:-** Data Wrangling

* Select Required Columns

```python
df = df[['sentence', 'subj_start', 'subj_end', 'obj_start', 'obj_end', 'relation']]
```

* Convert Relation to Category

```python
df['relation'] = df['relation'].astype('category')
```

* Filter Data

```python
df = df[df['sentence'].str.len() > 10]
```

**Explanation:**
* Keeps only important columns
* Converts categorical data
* Removes very short sentences (noise)
